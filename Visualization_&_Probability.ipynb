```
import pandas as pd

data = {
    "name": ["Ali", "Ali", "Sara", "Sara", "Rahul", "Rahul", "Anita", "Anita"],
    "subject": ["Math", "Science", "Math", "Science", "Math", "Science", "Math", "Science"],
    "score": [78, 85, 92, 88, 65, 70, 95, 90]
}

df = pd.DataFrame(data)
df.to_csv("students_data.csv", index=False)

print("CSV file created")
```

```
import pandas as pd

df = pd.read_csv("students_data.csv")
df
```

```
avg_scores = df.groupby("subject")["score"].mean()

avg_scores.plot(kind="bar")
plt.title("Average Score by Subject")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.show()
```

```
df["score"].plot(kind="line")
plt.title("Score Trend")
plt.xlabel("Index")
plt.ylabel("Score")
plt.show()
```

```
probability = 30 / 100
print("Probability of passing:", probability)
```

The probability of passing is 0.3, which means 30% of students passed. This indicates that the success rate is relatively low and improvements may be needed
